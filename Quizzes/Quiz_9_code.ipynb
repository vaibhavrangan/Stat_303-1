# Solution

## Q1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

midterm_data = pd.read_csv("/Users/vaibhavrangan/Downloads/Stat_303-1/Data/students_midterm_anonymous.csv")

wcas_mean = midterm_data[midterm_data["Program and Plan"].str.contains("Weinberg College of Arts & Sci")]["Midterm"].mean()
rest_mean = midterm_data[~midterm_data["Program and Plan"].str.contains("Weinberg College of Arts & Sci")]["Midterm"].mean()

print(wcas_mean > rest_mean)

```

## Q2

```python
ds_major_mean = midterm_data[midterm_data["Program and Plan"].str.contains("Data Science")]["Midterm"].mean()
non_major_mean = midterm_data[~midterm_data["Program and Plan"].str.contains("Data Science")]["Midterm"].mean()

# bar plot for the two means
plt.bar(["Data Science", "Non-Data Science"], [ds_major_mean, non_major_mean])
plt.xlabel("Major")
plt.ylabel("Mean Midterm Score")
plt.show()

print(ds_major_mean > non_major_mean)
```

```python
multiple_majors = midterm_data[midterm_data["Program and Plan"].str.count('/').ge(2)]["Midterm"].mean()
single_major = midterm_data[~midterm_data["Program and Plan"].str.count('/').ge(2)]["Midterm"].mean()

print(multiple_majors > single_major)
```

```python
correlation = midterm_data["Practice Midterm"].corr(midterm_data["Midterm"])

print(correlation)
```

```python
# Group by 'Level' and calculate the mean midterm score for each class year
mean_scores = midterm_data.groupby("Level")["Midterm"].mean()

# Plot the mean scores as a bar chart
mean_scores.plot(kind="bar")

# Add labels and title
plt.xlabel("Class Year")
plt.ylabel("Mean Midterm Score")
plt.title("Mean Midterm Scores by Class Year")
plt.show()
```
