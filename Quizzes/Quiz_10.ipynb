```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Food_Ingredients = pd.read_csv('/Users/vaibhavrangan/Downloads/Stat_303-1/Data/Food_Ingredients.csv')
Ingredient_prices = pd.read_csv('/Users/vaibhavrangan/Downloads/Stat_303-1/Data/Ingredient_prices.csv')
```

# Solution

## Q1

```python
# 100x6 np array 100 food items 6 ingredients
food_units = Food_Ingredients.iloc[:, :6].values
# 32x6 np array 32 grocery stores 6 ingredients
prices = Ingredient_prices.iloc[:, 1:].values

# transpose food_units
food_units = food_units.T

# transpose prices with food_units to get matrix of 32 grocery stores and 100 food items
total_cost = np.matmul(prices, food_units)
# sum across columns to find total cost per store
cost_per_store = total_cost.sum(axis=1)

# find index of store with minimum total cost, add 1 to go from index to store number
cheapest_store = np.argmin(cost_per_store) + 1
print(cheapest_store)
```

# Q2

```python
# apply np.var() to each column of total_cost to get variance for each food item
variances = np.var(total_cost, axis=0)
# find index of food item with highest variance
most_varied_food_item = np.argmax(variances) + 1
print(most_varied_food_item)
```

```python
insurance_df = pd.read_csv('/Users/vaibhavrangan/Downloads/Stat_303-1/Data/insurance_data.csv')
```

There are 3 categorical variables in the dataset, convert it to numerical values based on the following mapping dictionary using the `map()` function

```python
map_smoker = {'yes':1, 'no':0}
map_sex = {'male':0, 'female':1}
map_region = {'southeast':0, 'southwest':1, 'northwest':2, 'northeast':3}
```

## Q4

```python
insurance_df["smoker"] = insurance_df["smoker"].map(map_smoker)
insurance_df["sex"] = insurance_df["sex"].map(map_sex)
insurance_df["region"] = insurance_df["region"].map(map_region)
```

```python
# create a scatterplot of insurance costs on y-axis against BMI on x-axis
sns.scatterplot(data=insurance_df, x="bmi", y="charges");
```
